Calculate Relative Strength Index (RSI) and Bollinger Bands indicators that will define conditions to enter a trade. Using typical for mean-reversion strategy period of 14 days. 

```python
#Bollinger Bands loopback
ma_20 = 20

#Calculate Bollinger Bands 
df['ma_20'] = df['Adj Close'].rolling(period).mean()
df['std'] = df['Adj Close'].rolling(period).std()
df['upper_bollinger'] = df['ma_20'] + (2 * df['std'])
df['lower_bollinger'] = df['ma_20'] - (2 * df['std'])

#RSI loopback
period = 14

def calculate_rsi(prices: pd.Series, loopback: Callable):
    
    price_delta = prices.diff()
    price_delta = price_delta[1:] 
    
    gain, loss = price_delta.clip(lower=0), price_delta.clip(upper=0).abs()
    
    loopback_gain = loopback(gain)
    loopback_loss = loopback(loss)
    
    rs = loopback_gain/loopback_loss
    rsi = 100 - (100/(rs+1))
    
    return rsi

rsi_ema = calculate_rsi(df['Adj Close'], lambda s: s.ewm(span=period).mean()) #exponential weighted moving average
rsi_sma = calculate_rsi(df['Adj Close'], lambda s: s.rolling(period).mean())#simple moving average
rsi_rma = calculate_rsi(df['Adj Close'], lambda s: s.ewm(alpha=1 / period).mean()) #modified/running moving average

```
